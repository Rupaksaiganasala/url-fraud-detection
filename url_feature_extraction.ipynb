Feature Extraction

In this step, features are extracted from the URLs dataset.







```python
# Selecting relevant columns based on the requirements
selected_features = [
    #Dataset attributes based on URL.
    'length_url','qty_dot_url', 'qty_hyphen_url', 'qty_slash_url',
    'qty_questionmark_url', 'qty_equal_url', 'qty_at_url','email_in_url',

    #Dataset attributes based on domain URL.
    'domain_length','domain_in_ip','qty_dot_domain','qty_hyphen_domain',

    #Dataset attributes based on URL directory.
    'directory_length','qty_slash_directory',

    #Dataset attributes based on URL file name.
    'file_length','qty_dot_file','qty_hyphen_file',

    #Dataset attributes based on URL parameters.
    'params_length','qty_params',

    #Dataset attributes based on resolving URL and external services.
    'url_google_index','tls_ssl_certificate','qty_redirects','url_shortened',
    #Target variable
    'phishing'
]
```

```python
# Create a sub-dataset with the selected columns
sub_dataset = data[selected_features]
```

```python
# Save the sub-dataset to a CSV file
sub_dataset.to_csv('sub_dataset.csv', index=False)
```
